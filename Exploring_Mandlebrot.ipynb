This is to plot the mandlebrot set. For more information read here https://en.wikipedia.org/wiki/Mandelbrot_set

Import required libraries

```
import io
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
from PIL import Image 
from numpy import complex, array 
```

Function to compute whether the point x + iy lies in the Mandlebrot set or not. Calculate Z_n+1 = Z_n + c where c = x + iy. If the series explodes then the point is outside the set and if its magnitude remains less than 2 then it is withing the set.

```
def checkmandlebrot(x,y):
  c = complex(x,y)
  maxiter = 1000
  z = 0
  for i in range(maxiter):
    if( abs(z) > 2):
      return ( min(int(i%255)+80, 255), 0, 0)
    z = z*z + c
  return (0,0,0)
```

Creating an image from scratch and fill the pixels with color based on whether the point is within the set or not. The adjustments are to ensure that that  the orgin shows up at the centre of the image

```
width = 1024
img = Image.new('RGB', (width, width)) 
pixels = img.load() 

img.width
img.height
for i in range(1,img.width):
  for j in range(1,img.height):
    x = (i - 3/4*width)*4/width
    y = (j - 1/2*width)*4/width
    pixels[i,j] = checkmandlebrot( x,y )
```

Plot the beautiful Mandlebrot set

```
plt.figure(figsize = (10,10))
plt.imshow(img)
```
